```python
# Import Pandas
import pandas as pd

# Import seaborn
import warnings # we'll ignore warnings
warnings.filterwarnings("ignore")

import seaborn as sns

import matplotlib.pyplot as plt
sns.set(style="white", color_codes=True)
```

```python
# Lets load the the Iris flower dataset
iris = pd.read_csv('Desktop/Data/Iris.csv') # this is a Pandas DF now
```

```python
# Let's explore the data
iris.head()
```

```python
# Let's see how many examples we have of each species
iris["Species"].value_counts()
```

```python
# Scatterplots - Sepal length vs Sepal width
iris.plot(kind="scatter", x="SepalLengthCm", y="SepalWidthCm")
```

```python
# Similar plot using Seaborn
# A seaborn jointplot shows bivariate scatterplots and univariate histograms in the same figure
sns.jointplot(x="SepalLengthCm", y="SepalWidthCm", data=iris, size=5)
```

```python
# Scatterplots - PetalLengthCm vs PetalWidthCm
iris.plot(kind="scatter", x="PetalLengthCm", y="PetalWidthCm")
```

```python
# Similar plot using Seaborn
# A seaborn jointplot shows bivariate scatterplots and univariate histograms in the same figure
sns.jointplot(x="PetalLengthCm", y="PetalWidthCm", data=iris, size=5)
```

```python
# We'll use seaborn's FacetGrid to Multi-plot grid for plotting conditional relationships. We are using scatteplots differentiated by species.This gives a different color to the points in the scatter plot based on their Species value
sns.FacetGrid(iris, height=5,hue="Species").map(plt.scatter,"SepalLengthCm","SepalWidthCm").add_legend()
```

```python
sns.FacetGrid(iris, height=5,hue="Species").map(plt.scatter,"PetalLengthCm","PetalWidthCm").add_legend()
```

```python
# boxplot shows the distribution of quantitative data
sns.boxplot(x="Species", y="PetalLengthCm", data=iris)
```

```python
#A stripplot with jitter helps to visualize the distribution of values 
# Lets use jitter=True so that all the points don't fall in single vertical line
abc = sns.boxplot(x="Species", y="PetalLengthCm", data=iris)
abc = sns.stripplot(x="Species", y="PetalLengthCm", data=iris, jitter=True, edgecolor="gray")
```

```python
abc = sns.boxplot(x="Species", y="PetalLengthCm", data=iris)
abc = sns.stripplot(x="Species", y="PetalLengthCm", data=iris, jitter=True, edgecolor="gray")
```

```python
abc = sns.boxplot(x="Species", y="SepalLengthCm", data=iris)
abc = sns.stripplot(x="Species", y="SepalLengthCm", data=iris, jitter=True, edgecolor="gray")
```

```python
abc = sns.boxplot(x="Species", y="SepalWidthCm", data=iris)
abc = sns.stripplot(x="Species", y="SepalWidthCm", data=iris, jitter=True, edgecolor="gray")
```

```python
# A violin plot combines scater plot and boxplot.We can see the densitity of the distribution of the values.
# Denser regions of the data are fatter, and sparser thiner in a violin plot
plt.figure(figsize=(15,10))
plt.subplot(2,2,1)
sns.violinplot(x='Species',y='PetalLengthCm',data=iris)
plt.subplot(2,2,2)
sns.violinplot(x='Species',y='PetalWidthCm',data=iris)
plt.subplot(2,2,3)
sns.violinplot(x='Species',y='SepalLengthCm',data=iris)
plt.subplot(2,2,4)
sns.violinplot(x='Species',y='SepalWidthCm',data=iris)
```

```python
#A kernel density estimate (KDE) plot is a method for visualizing the distribution of observations in a dataset, analogous to a histogram. KDE represents the data using a continuous probability density curve in one or more dimensions.Relative to a histogram, KDE can produce a plot that is less cluttered and more interpretable, especially when drawing multiple distributions
```

```python

sns.FacetGrid(iris, hue="Species", height=6).map(sns.kdeplot, "PetalLengthCm").add_legend()
```

```python
sns.FacetGrid(iris, hue="Species", height=6).map(sns.kdeplot, "PetalWidthCm").add_legend()
```

```python
sns.FacetGrid(iris, hue="Species", height=6).map(sns.kdeplot, "SepalLengthCm").add_legend()
```

```python
sns.FacetGrid(iris, hue="Species", height=6).map(sns.kdeplot, "PetalLengthCm").add_legend()
```

```python
# Pairplot: which shows the bivariate relation between each pair of features
# Iris-setosa species is very well separataed from the other two across all feature combinations
sns.pairplot(iris.drop("Id", axis=1), hue="Species", size=3)
```

```python
#  The diagonal elements in a pairplot show the histogram by default in above chart. # Lets look into the univariate distributions more closely. Lets use kdeplot,KDE: Kernel Density Estimate.
sns.pairplot(iris.drop("Id", axis=1), hue="Species", size=3, diag_kind="kde")
```

```python
# Pandas boxplot
iris.drop("Id", axis=1).boxplot(by="Species", figsize=(12, 6))
```

```python
# One cool more sophisticated technique pandas has available is called Andrews Curves
# Andrews Curves involve using attributes of samples as coefficients for Fourier series
# and then plotting these
from pandas.plotting import andrews_curves
andrews_curves(iris.drop("Id", axis=1), "Species")
```

```python
# Another multivariate visualization technique pandas has is parallel_coordinates
# Parallel coordinates plots each feature on a separate column & then draws lines
# connecting the features for each data sample
from pandas.plotting import parallel_coordinates
parallel_coordinates(iris.drop("Id", axis=1), "Species")
```

```python
# pandas has a multivariate visualization tecnique-radviz.
#RadViz is useful to plot multidimensional data in 2D
from pandas.plotting import radviz
radviz(iris.drop("Id", axis=1), "Species")
```
