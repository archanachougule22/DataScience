**Objective**
write a code to merge the tables Data1 & Data2.

Lets import pandas library

```python
import pandas as pd
```

Read the datasets data1 & data2. Check if both datasets are importted correctly.

```python
# Read data from Data 1 table
data1 = pd.read_csv(Desktop/Test_folder/Data1.csv')
print(data1)
```

```python
# Read data from Data 2 table
data2 = pd.read_csv(Desktop/Test_folder/Data2.csv')
print(data2)
```

We observe that the date portion of the "Data1" and "Data 2" columns follows the format "Test1_MMYY" and "Test2_MMYY" respectively. Lets split the string using "_" as the delimiter and extracts the month-year .Convert the string into pd.to_datetime() with the format "%m%Y". T

```python
# Extract month-year from Data 1
data1["Month_Year"] = data1["Data1"].str.split("_").str[1]
data1['Month_Year'] = pd.to_datetime(data1['Month_Year'],format="%m%Y").dt.strftime("%m-%Y")
print(data1)
```

```python
# Extract month-year from Data 2
data2["Month_Year"] = data2["Data2"].str.split("_").str[1]
data2['Month_Year'] = pd.to_datetime(data2['Month_Year'],format="%m%Y").dt.strftime("%m-%Y")
print(data2)
```

Lets Merge the results with Customer_id & MonthYear.Outer join is used as some of the observations can be present only in 1 dataset.
Results are sorted by 'Customer_id ' and 'Month Year'


```python
# Join data for same customer ID and month-year
merged_data = pd.merge(data1, data2, how="outer", left_on=["Customer_id", "Month_Year"], right_on=["Customer_id", "Month_Year"])
# Change the order of columns
merged_data = merged_data[["Customer_id", "Month_Year", "Data1", "Data2"]]
# Sort the data by Customer_id & Month_Year
merged_data = merged_data.sort_values(by=["Customer_id", "Month_Year"])
#Check the results
print(merged_data)
```

We can see that tables are merged properly.
